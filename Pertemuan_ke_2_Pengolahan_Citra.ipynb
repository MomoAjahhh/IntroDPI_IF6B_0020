```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_bgr = cv2.imread('Araka.jpg')

img_bgr = cv2.resize(img_bgr, (400, 400))

print("Tipe data:", type(img_bgr))
print("Shape    :", img_bgr.shape)
print("Dtype    :", img_bgr.dtype)

print()
print("Artinya:")
print(f" Tinggi  = {img_bgr.shape[0]} piksel")
print(f" Lebar   = {img_bgr.shape[1]} piksel")
print(f" Channel = {img_bgr.shape[2]} (BGR)")

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

print("Total Piksel: ", img_bgr.shape[0] * img_bgr.shape[1])

plt.figure(figsize=(4,6))
plt.imshow(img_rgb)
plt.title('Gambar Wajah Araka\n(sudah dikonversi BGR -> RGB)')
plt.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

area_kecil = img_gray[200:208, 180:188]

print("=== MATRIKS f(x,y) - Area 8x8 Piksel ===")
print("Setiap angka = nilai intensitas 1 piksel (0=hitam, 255=putih)")
print()
print(area_kecil)
print()
print("Contoh baca:", f"f(180, 200) = {img_gray[200,180]}")
print("Contoh baca:", f"f(181, 201) = {img_gray[201,181]}")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(img_gray, cmap='gray', vmin=0, vmax=255)
axes[0].set_title('Gambar Grayscale\n(setiap piksel = 1 angka)')
axes[0].axis('off')

import matplotlib.patches as patches
rect = patches.Rectangle((180, 200), 8, 8, linewidth=2, edgecolor='red', facecolor='none')
axes[1].add_patch(rect)
axes[0].text(190, 195, 'area ini\ndizoom', color='red', fontsize=8)

axes[1].imshow(area_kecil, cmap='gray', vmin=0, vmax=255)
for i in range(8):
  for j in range(8):
    nilai = area_kecil[i, j]
    warna_teks = 'white' if nilai < 123 else 'black'
    axes[1].text(j, i, str(nilai), ha='center', va='center',
                 fontsize=9, color=warna_teks, fontweight='bold')
axes[1].set_title('Zoom 8x8 Piksel\nAngka =nilai f(x,y)')
axes[1].axis('off')

plt.suptitle('SLIDE 4: Representasi Matematis f(x,y)', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

area_kecil_rgb = img_rgb[200:208, 180:188]

print("=== MATRIKS f(x,y) - Area 8x8 Piksel (RGB) ===")
print("Setiap angka = nilai intensitas 1 piksel (R,G,B)")
print()
print(area_kecil_rgb)
print()

print("Contoh baca:", f"f(180, 200) = {img_rgb[200,180]}")
print("Contoh baca:", f"f(181, 201) = {img_rgb[201,181]}")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(img_rgb)
axes[0].set_title('Gambar RGB Keseluruhan')
axes[0].axis('off')

import matplotlib.patches as patches

rect = patches.Rectangle((180, 200), 8, 8, linewidth=2, edgecolor='red', facecolor='none')
axes[0].add_patch(rect)
axes[0].text(190, 195, 'area ini\ndizoom', color='red', fontsize=8)


axes[1].imshow(area_kecil_rgb)
for i in range(8):
  for j in range(8):
    rgb_nilai = area_kecil_rgb[i, j]
    kecerahan_piksel = np.mean(rgb_nilai)
    warna_teks = 'white' if kecerahan_piksel < 123 else 'black'
    r, g, b, = rgb_nilai
    axes[1].text(j, i, f"{r}\n{g}\n{b}", ha='center', va='center',
                 fontsize=8, color=warna_teks, fontweight='bold')
axes[1].set_title('Zoom 8x8 Piksel (RGB)\nAngka = nilai f(x,y)')
axes[1].axis('off')

plt.suptitle('SLIDE 4: Representasi Matematis f(x,y) (RGB)', fontweight='bold')
plt.tight_layout()

ukuran_bytes = img_rgb.nbytes
total_mb = ukuran_bytes / (1024 * 1024)

print(f"\nUkuran gambar RGB: {total_mb:.2f} MB")
print()
plt.show()
```
